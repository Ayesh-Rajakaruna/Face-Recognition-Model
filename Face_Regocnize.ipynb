```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline 
import os
```

```python
Input_Folder_Name = "./InputData/"
Output_Folder_Name = "./OutputData/"
```

```python
face_cascade = cv.CascadeClassifier("Sourse//Face.xml")
eye_cascade = cv.CascadeClassifier("Sourse//Eye.xml")
count = 1
```

```python
def Detect_Fase(img_gray, count):
    for face in face_cascade.detectMultiScale(img_gray, 1.3, 5):
        x, y, w, h = face
        face_gray = img_gray[y:y+h,x:x+w]
        number_of_eye = len(eye_cascade.detectMultiScale(face_gray, 1.3, 5))
        if number_of_eye == 2:
            filename = os.path.join(Output_Folder_Name,str(count)+".jpg")
            cv.imwrite(filename, face_gray)
            count += 1
    return count
```

```python
for Path, Folders, Files in os.walk(Input_Folder_Name):
    for i in Files:
        Image_Parth = os.path.join(Input_Folder_Name,i)
        print(Image_Parth)
        Img = cv.imread(Image_Parth)
        Img_gray = cv.cvtColor(Img, cv.COLOR_BGR2GRAY)
        count = Detect_Fase(Img_gray, count)
        
```
